# Logistic Regression with Synthetic Data
## Validation Exercises

With each of the validation metrics, the metric will be done on both the training set and a randomly generated testing set.

```python
import numpy.random # for generating our dataset
from sklearn import linear_model # for fitting our model

from sklearn.metrics import roc_curve
from sklearn.metrics import auc

# force numpy not to use scientific notation, to make it easier to read the numbers the program prints out
numpy.set_printoptions(suppress=True)

# to display graphs in this notebook
%matplotlib inline
import matplotlib.pyplot
```

```python
NUM_INPUTS = 50 # inputs per class
PLANT_A_AVG_HEIGHT = 60.0
PLANT_A_AVG_WIDTH = 9.0
PLANT_B_AVG_HEIGHT = 59.0
PLANT_B_AVG_WIDTH = 10.0

# Pick numbers randomly with a normal distribution centered around the averages

plant_a_heights = numpy.random.normal(loc=PLANT_A_AVG_HEIGHT, size=NUM_INPUTS)
plant_a_widths = numpy.random.normal(loc=PLANT_A_AVG_WIDTH, size=NUM_INPUTS)

plant_b_heights = numpy.random.normal(loc=PLANT_B_AVG_HEIGHT, size=NUM_INPUTS)
plant_b_widths = numpy.random.normal(loc=PLANT_B_AVG_WIDTH, size=NUM_INPUTS)

# this creates a 2-dimensional matrix, with heights in the first column and widths in the second
# the first half of rows are all plants of type a and the second half are type b
plant_inputs = list(zip(numpy.append(plant_a_heights, plant_b_heights),
                        numpy.append(plant_a_widths, plant_b_widths)))

# this is a list where the first half are 0s (representing plants of type a) and the second half are 1s (type b)
classes = [0]*NUM_INPUTS + [1]*NUM_INPUTS
```

```python
model = linear_model.LogisticRegression()
model.fit(plant_inputs, classes)

print('Intercept: {0}  Coefficients: {1}'.format(model.intercept_, model.coef_))
```

## Visualization

```python
def plot_plants():
    # create a figure and label it
    fig = matplotlib.pyplot.figure()
    fig.suptitle('Plant Data Set')
    matplotlib.pyplot.xlabel('Height')
    matplotlib.pyplot.ylabel('Width')

    # put the generated points on the graph
    a_scatter = matplotlib.pyplot.scatter(plant_a_heights, plant_a_widths, c="red", marker="o", label='plant a')
    b_scatter = matplotlib.pyplot.scatter(plant_b_heights, plant_b_widths, c="blue", marker="^", label='plant b')

    # add a legend to explain which points are which
    matplotlib.pyplot.legend(handles=[a_scatter, b_scatter])

def division_line(x,model):
    # Get coefficients
    coef=model.coef_
    inter=model.intercept_
    
    # 1/(1+exp(ax+by+c))=0.5 <=> exp(ax+by+c)=1 <=> ax+by+c=0
    y=-1/coef[0,1]*(coef[0,0]*x+inter[0])
    return(y)

def plot_with_line(model):
    plot_plants()

    # Getting max and mins
    mins=[min(numpy.append(plant_a_heights,plant_b_heights)), min(numpy.append(plant_a_widths, plant_b_widths))]
    maxs=[max(numpy.append(plant_a_heights,plant_b_heights)), max(numpy.append(plant_a_widths, plant_b_widths))]

    # Making division line
    X1=numpy.linspace(mins[0],maxs[0])
    Y1=division_line(X1,model)

    matplotlib.pyplot.plot(X1, Y1, label='Division Line')

    # Update legend
    matplotlib.pyplot.legend()

    # Display plot
    matplotlib.pyplot.show()
    
plot_with_line(model)
```

## Model Validation 1: F-Measure
Let's validate the model with an F-Measure function. Here, a value of 1 indicates perfect prediction, while a value of 0.5 would be random chance. A value of 0 is perfectly incorrect prediction (which is perfect prediction in its own way). In this case, the model is getting somewhat bad scores (0.7-0.8 range); this is likely because I made the average values for plant A and B rather close, so as to make the model's task more difficult. Also, it is worth noting that the model distinguishes the training set better than the test set, though this isn't really too surprising given that the model will naturally fit better onto its own training set.

```python
def F_measure(model,data, targets):
    predictions=model.predict(data)
    
    true_positives=[sum([prediction==target==i for prediction, target in zip(predictions,targets)]) for i in model.classes_]
    false_negatives=[sum([prediction!=target==i for prediction, target in zip(predictions,targets)]) for i in model.classes_]
    false_positives=[sum([prediction==i!=target for prediction, target in zip(predictions,targets)]) for i in model.classes_]
    
    precision=[a/(a+b) for a,b in zip(true_positives, false_positives)]
    recall=[a/(a+b) for a,b in zip(true_positives, false_negatives)]
    f_meas=[2*(P*R)/(P+R) for P,R in zip(precision, recall)]
    
    return(f_meas)

test_A=[[numpy.random.normal(loc=PLANT_A_AVG_HEIGHT),numpy.random.normal(loc=PLANT_A_AVG_WIDTH)] for i in range(50)]
test_B=[[numpy.random.normal(loc=PLANT_B_AVG_HEIGHT),numpy.random.normal(loc=PLANT_B_AVG_WIDTH)] for i in range(50)]

test_inputs=test_A+test_B

print(F_measure(model,plant_inputs,classes))
print(F_measure(model,test_inputs,classes))
```

## Model Validation 2: ROC Curve & AUC

An AUC value of 1 would indicate the modle, if given a sample from class A and B, would always assign a higher probability of the sample from class A belonging to class A (and similarly for class B); a value of 0 would indicate it never does this. A value of 0.5 corresponds to a random model. In this case, the model is getting scores of about 0.85 on the training set and 0.75 on the test set, which can likely be explained in a similar way as above (the averages of the set are pretty close, and the model should be doing better on the training set).

## Model Validation 3: Precision-Recall Curve

The precision-recall curve seems more qualitative than the previous two, but from what I can guess a good "score" would be as flat of a PR-curve as possible. As above, the model doesn't really have flat PR-curves, likely due to the similarity of the two classes it is trying to distinguish. Qualitatively, the training set's PR-curve looks somewhat flatter.

```python
def PR_plot_point(prob, model, Inputs):
    predictions=[[int(i>=prob) for i in model.predict_proba([inputs])[0]] for inputs in Inputs]
    
    true_positives=sum([prediction[truth]==1 for prediction, truth in zip(predictions, classes)])
    
    false_positives=sum([prediction[1-truth]==1 for prediction, truth in zip(predictions, classes)])
    false_negatives=sum([prediction[truth]==0 for prediction, truth in zip(predictions, classes)])
    
    recall=true_positives/(true_positives+false_negatives)
    if true_positives+false_positives==0:
        precision=1
    else:
        precision=true_positives/(true_positives+false_positives)
    return([recall, precision])
    
    
def PR_plot(model, Inputs):
    fig = matplotlib.pyplot.figure()
    fig.suptitle('Precision-Recall Curve')
    matplotlib.pyplot.xlabel('Recall')
    matplotlib.pyplot.ylabel('Precision')
    
    X=numpy.linspace(0,1, 200)
    points=numpy.array([PR_plot_point(prob,model, Inputs) for prob in X])
    
    matplotlib.pyplot.plot(points[:,0], points[:,1], c='black', label=f'PR Curve')
    
    matplotlib.pyplot.show()
    

PR_plot(model, plant_inputs)    
PR_plot(model, test_inputs)
```
